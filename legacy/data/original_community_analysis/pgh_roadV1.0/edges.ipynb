```python
import networkx as nx
import osmnx as ox
import requests
import matplotlib.cm as cm
import matplotlib.colors as colors

%matplotlib inline
ox.config(log_file=False, log_console=True, use_cache=True)
```

```python
# get a graph for some city
pgh = ox.graph_from_place('Pittsburgh, Pennsylvania, USA', network_type='drive', simplify=True, buffer_dist=2000)
pgh_prj = ox.project_graph(pgh)
fig, ax = ox.plot_graph(pgh_prj, save=True, file_format='svg', filename="pgh_road")
```

```python
# Output to shp
ox.save_graph_shapefile(pgh_prj, filename='pgh_road')
```

```python
# Output for NetworkX
ox.save_graphml(pgh_prj, filename='pgh_road.graphml',gephi=True)
```
